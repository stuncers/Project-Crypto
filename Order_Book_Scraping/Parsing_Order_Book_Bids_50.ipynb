```python
import sched, time
import numpy as np
import pandas as pd
import datetime
s = sched.scheduler(time.time, time.sleep)
```

```python

def do_something(sc): 
    element = driver.find_element(By.CLASS_NAME, "orderlist-container")

    # Extract the text from the div element
    data = element.text
    x = data.split("\n")
    #asks= x[:[ i for i, word in enumerate(x) if word.endswith('More') ][0]-2]
    #bids= x[[ i for i, word in enumerate(x) if word.endswith('More') ][0]+1:]
    df_bids = pd.DataFrame(np.array(x).reshape(-1,3), columns=['Price(USDT)', 'Amount(BTC)', 'Total'])
    #df_asks = df_asks.applymap(lambda x: x.replace(',', '') if isinstance(x, str) else x)
    #df_bids = pd.DataFrame(np.array(bids).reshape(-1,3), columns=['Price(USDT)', 'Amount(BTC)', 'Total'])
    df_bids = df_bids.applymap(lambda x: x.replace(',', '') if isinstance(x, str) else x)
    #df_asks['Total']=df_asks.Total.astype('int64')
    #df_asks['Price(USDT)']=df_asks['Price(USDT)'].astype('int64')
    #df_asks['Amount(BTC)']=df_asks['Amount(BTC)'].astype('float64')
    df_bids['actual']= df_bids['Price(USDT)'].iloc[0]
    df_bids=df_bids.iloc[1: , :]
    df_bids['Total']=df_bids.Total.astype('float64')
    df_bids['Price(USDT)']=df_bids['Price(USDT)'].astype('float64')
    df_bids['Amount(BTC)']=df_bids['Amount(BTC)'].astype('float64')
    df_bids['time']= datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    df_bids['time']=pd.to_datetime(df_bids['time'], format="%Y-%m-%d %H:%M:%S")
    #df_bids['time'] = df_bids['time'].apply(lambda x: x.replace(second=59))
    #df_asks['time']= datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    #df_asks['time']=pd.to_datetime(df_asks['time'], format="%Y-%m-%d %H:%M:%S")
    #df_asks['time'] = df_asks['time'].apply(lambda x: x.replace(second=59))
    #df_asks_100=pd.read_csv('/cta/users/tuna/jupyter/Scraping_Order_Book/Order_Data/Scrape/asks_100.csv')
    #df_asks_100= df_asks_100.drop('Unnamed: 0',axis=1)
    #df_asks_100 = pd.concat([df_asks_100,df_asks])
    #df_asks_100.to_csv('/cta/users/tuna/jupyter/Scraping_Order_Book/Order_Data/Scrape/asks_100.csv')
    df_bids_100=pd.read_parquet('/cta/users/tuna/jupyter/Scraping_Order_Book/Order_Data/Scrape/bids_50.parquet', engine='pyarrow')
    #df_bids_100= df_bids_100.drop('Unnamed: 0',axis=1)
    df_bids_100 = pd.concat([df_bids_100,df_bids])
    df_bids_100.to_parquet('/cta/users/tuna/jupyter/Scraping_Order_Book/Order_Data/Scrape/bids_50.parquet', engine='pyarrow')
    sc.enter(.1, 1, do_something, (sc,))
    return print('Jobs done successfully!, Returning to Start',
                 'Total length of DF', len(df_bids_100))
    
#arrange the weight of prices and careful with the differentiation
s.enter(.1, 1, do_something, (s,))
s.run()
    
```
